```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/pm25_Reno_2024-07-01.csv",
                 parse_dates=["datetime_utc"])

plt.figure(figsize=(10,4))
plt.plot(df["datetime_utc"], df["pm25"], lw=0.7)
plt.title("Hourly PM₂.₅ in Reno, July 2024")
plt.ylabel("µg/m³")
plt.xlabel("UTC time")
plt.tight_layout()
plt.show()
```

```python
daily = (df.set_index("datetime_utc")
           .resample("1D")["pm25"]
           .median())        # median is more robust than mean

plt.figure(figsize=(6,4))
plt.boxplot(daily.dropna(), vert=True, labels=["July"])
plt.title("Daily median PM₂.₅ – variability")
plt.ylabel("µg/m³")
plt.show()
```
